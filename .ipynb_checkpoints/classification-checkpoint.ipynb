```python
%matplotlib inline
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Rescaling, CategoryEncoding, Conv2D, MaxPooling2D, Dense, Flatten, Dropout
from tensorflow.keras.metrics import Precision, Recall
import tensorflow as tf
import os
import cv2
```

```python
num = list(range(6))
num_left = list(map(lambda x:str(x)+'L_label', num))
num_right = list(map(lambda x:str(x)+'R_label', num))
```

```python
def extract_label(base):
    for filename in os.listdir(base):
        if filename.split('.')[0][-2:] == '0L': 
          os.rename(base+'/'+filename, base+'/'+'0L_label'+'/'+filename)
        if filename.split('.')[0][-2:] == '1L': 
          os.rename(base+'/'+filename, base+'/'+'1L_label'+'/'+filename)
        if filename.split('.')[0][-2:] == '2L': 
          os.rename(base+'/'+filename, base+'/'+'2L_label'+'/'+filename)
        if filename.split('.')[0][-2:] == '3L': 
          os.rename(base+'/'+filename, base+'/'+'3L_label'+'/'+filename)
        if filename.split('.')[0][-2:] == '4L': 
          os.rename(base+'/'+filename, base+'/'+'4L_label'+'/'+filename)
        if filename.split('.')[0][-2:] == '5L': 
          os.rename(base+'/'+filename, base+'/'+'5L_label'+'/'+filename)
        if filename.split('.')[0][-2:] == '0R': 
          os.rename(base+'/'+filename, base+'/'+'0R_label'+'/'+filename)
        if filename.split('.')[0][-2:] == '1R': 
          os.rename(base+'/'+filename, base+'/'+'1R_label'+'/'+filename)
        if filename.split('.')[0][-2:] == '2R': 
          os.rename(base+'/'+filename, base+'/'+'2R_label'+'/'+filename)
        if filename.split('.')[0][-2:] == '3R': 
          os.rename(base+'/'+filename, base+'/'+'3R_label'+'/'+filename)
        if filename.split('.')[0][-2:] == '4R': 
          os.rename(base+'/'+filename, base+'/'+'4R_label'+'/'+filename)
        if filename.split('.')[0][-2:] == '5R': 
          os.rename(base+'/'+filename, base+'/'+'5R_label'+'/'+filename)
```

```python
train_base = 'C:/Users/jeff/Desktop/fingers/train'
test_base = 'C:/Users/jeff/Desktop/fingers/test'
#os.chdir(train_base)
#for label in num_left:
#  os.mkdir(label)
#for label in num_right:
#   os.mkdir(label)
#extract_label(train_base)

#os.chdir(test_base)
#for label in num_left:
#  os.mkdir(label)
#for label in num_right:
#    os.mkdir(label)
#extract_label(test_base)
```

```python
print('訓練集的每個特徵分別有{}筆資料'.format(len(os.listdir(train_base+'/'+'0R_label'))))
print('測試集的每個特徵分別有{}筆資料'.format(len(os.listdir(test_base+'/'+'0R_label'))))
```

```python
image_size = (128,128)
batch_size = 100

train_dataset = tf.keras.utils.image_dataset_from_directory(
    train_base,
    color_mode='grayscale',
    validation_split=0.2,
    subset='training',
    seed=2023,
    image_size=image_size,
    batch_size=batch_size)

val_dataset = tf.keras.utils.image_dataset_from_directory(
    train_base,
    color_mode='grayscale',
    validation_split=0.2,
    subset='validation',
    seed=2023,
    image_size=image_size,
    batch_size=batch_size)

test_dataset = tf.keras.utils.image_dataset_from_directory(
    test_base,
    color_mode='grayscale',
    seed=2023,
    image_size=image_size,
    batch_size=batch_size)
```

```python
data_iterator = train_dataset.as_numpy_iterator()
batch = data_iterator.next()
plt.figure(figsize = (8,4))
for idx, img in enumerate(batch[0][:10]):
  plt.subplot(2,5,idx+1)
  plt.imshow(img, cmap='gray')
  plt.title('{}'.format(batch[1][idx]))
  plt.axis('off')
```

### 標籤對應
    0=>0L, 2=>1L, 4=>2L, 6=>3L, 8=>4L, 10=>5L
    1=>0R, 3=>1R, 5=>2R, 7=>3R, 9=>4R, 11=>5R

```python
model = Sequential()

model.add(Rescaling(scale = 1.0 / 255))
model.add(Conv2D(16, (3,3), padding='same', activation='relu', input_shape=(128,128,1)))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Conv2D(32, (3,3), padding='same', activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Conv2D(64, (3,3), padding='same', activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Flatten())
model.add(Dense(64, activation='relu'))
model.add(Dense(12, activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(
    loss='sparse_categorical_crossentropy',
    optimizer='Adam',
    metrics=['accuracy'])
```

```python
model.fit(
    train_dataset,
    epochs=10,
    validation_data=val_dataset)
```

```python
loss, acc = model.evaluate(test_dataset)
print(f'測試資料的正確率為 {acc*100:.2f}%')
```
